```python
import pandas as pd
import matplotlib.pyplot as plt


country = input('Country: >> ')
start_year = int(input('Start Year: >> '))
end_year = int(input('End Year: >> '))


gdp_values = []
years = []


if country in gdp_data['Country Name'].values:
    print(f'GDP of {country} from {start_year} to {end_year}:')
    for year in range(start_year, end_year + 1):
        try:
            
            gdp_value = gdp_data[gdp_data['Country Name'] == country][str(year)].values[0]
            gdp_values.append(gdp_value)
            years.append(year)
            print(f'Year {year}: {gdp_value}')
        except:
            print(f'Year {year}: Not available')
else:
    print(f'{country} not found in the dataset')


if gdp_values:
    plt.bar(years, gdp_values, color='skyblue')
    plt.xlabel('Year')
    plt.ylabel('GDP')
    plt.title(f'GDP of {country} from {start_year} to {end_year}')
    plt.xticks(years, rotation=45)  
    plt.tight_layout()  
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


country1 = input('Country 1: >> ')
country2 = input('Country 2: >> ')
start_year = int(input('Start Year: >> '))
end_year = int(input('End Year: >> '))


gdp_values_1 = []
gdp_values_2 = []
years = []


if country1 in gdp_data['Country Name'].values and country2 in gdp_data['Country Name'].values:
    print(f'Comparing GDP of {country1} and {country2} from {start_year} to {end_year}:')
    
    for year in range(start_year, end_year + 1):
        try:
           
            gdp_value_1 = gdp_data[gdp_data['Country Name'] == country1][str(year)].values[0]
            gdp_value_2 = gdp_data[gdp_data['Country Name'] == country2][str(year)].values[0]
            
            gdp_values_1.append(gdp_value_1)
            gdp_values_2.append(gdp_value_2)
            years.append(year)
            
            print(f'Year {year}: {country1} = {gdp_value_1}, {country2} = {gdp_value_2}')
        except:
            print(f'Year {year}: Data not available for one or both countries')
else:
    print(f'One or both countries not found in the dataset')


if gdp_values_1 and gdp_values_2:
    
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax1.bar(years, gdp_values_1, width=0.4, label=f'{country1} GDP', color='skyblue', align='center')
    ax1.bar(years, gdp_values_2, width=0.4, label=f'{country2} GDP', color='lightgreen', align='edge')
    
    ax1.set_xlabel('Year')
    ax1.set_ylabel('GDP')
    ax1.set_title(f'GDP Comparison of {country1} and {country2} from {start_year} to {end_year}')
    ax1.set_xticks(years)
    ax1.set_xticklabels(years, rotation=45)
   
    ax2 = ax1.twinx()  
    ax2.plot(years, gdp_values_1, color='blue', marker='o', label=f'{country1} GDP (Line)', linestyle='--')
    ax2.plot(years, gdp_values_2, color='green', marker='o', label=f'{country2} GDP (Line)', linestyle='--')
    
    ax2.set_ylabel('GDP (Line)', color='black')
    
   
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    
    plt.tight_layout()  
    plt.show()
```

```python


def tukey_rule(data_frame, column_name):
    data = data_frame[column_name]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)

    IQR = Q3 - Q1

    max_value = Q3 + 1.5 * IQR
    min_value = Q1 - 1.5 * IQR
    
    return data_frame[(data_frame[column_name] < max_value) & (data_frame[column_name] > min_value)]
```

```python

df_outlier_removed = df_2016.copy()

for column in ['population','gdp']:
    df_outlier_removed = tukey_rule(df_outlier_removed, column)
```

```python
def normalize(x, x_min, x_max):
  
    return (x - x_min) / (x_max - x_min)
```

```python
class Normalizer():
   
    
    def __init__(self, dataframe):
        
        
        
        self.params = []

        for column in dataframe.columns:
            self.params.append(x_min_max(dataframe[column]))
            
    def x_min_max(data):
      
        minimum = min(data)
        maximum = max(data)
        return minimum, maximum

    def normalize_data(self, x):
     
        normalized = []
        for i, value in enumerate(x):
            x_max = self.params[i][1]
            x_min = self.params[i][0]
            normalized.append((x[i] - x_min) / (x_max - x_min))
        return normalized
```
